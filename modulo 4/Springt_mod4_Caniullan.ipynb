```python
# importar librerias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


# ejercicio 1, crear dataframe
data = {'days':['L', 'M', 'X', 'J', 'V'],'calorias':[420,380,390,490,300],'tiempo':[60,40,75,55,45]}
dataframe = pd.DataFrame(data)
df=dataframe.set_index('days')
df
```

```python
# Ejercicio 2, calcular la media, mediana, y desviación típica de ambas columnas.
print('Medidas de tendencia cenral de la variable calorias ')
print('media: ', df['calorias'].mean())
print('mediana: ', df['calorias'].median())
print('desviacion tipica: ', df['calorias'].std())

print('*'*50)

print('Medidas de tendencia cenral de la variable tiempo ')
print('media: ', df['tiempo'].mean())
print('mediana: ', df['tiempo'].median())
print('desviacion tipica: ', df['tiempo'].std())
```

```python
# ejercicio 3, agregar columna booleana si calorias es mayor a 400
dfbool = df['calorias'] > 400
df['reto'] = dfbool
df
```

```python
# Ejercicio 4, filtrar el DataFrame y devolver otro DataFrame con las filas pares que
# cumplan que el número de calorías es mayor de 400.

df2 = df['calorias'][::2]
df3 = df2[df2 > 400]
df3 = pd.DataFrame(df3)
df3
```

```python
# Ejercicio 5, crear a partir del DataFrame una serie con los porcentajes de días que
# se ha conseguido el reto y los que no.

porc = df['reto'].value_counts()*100/5 
print(porc) 
```

```python
# Ejercicio 6, crear un gráfico como el de más abajo que muestre la progresión de
# las calorías y tiempo durante la semana.

df.plot.line(y=['calorias', 'tiempo'],figsize=(12,3),lw=1);
```
