```python
# importar librerias
import pandas as pd
import numpy as np
import seaborn as sns

# cargar dataframe titanic
df = pd.read_csv('/content/titanic3.csv',delimiter = ';')
df
```

```python
# Grafico de barras % hombre y mujeres sobrevivientes
sns.barplot(x='sex', y='survived', data=df)
```
