```python
# importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# cargar dataframe
df = pd.read_csv('/content/M4-AP03_casasboston.csv')
df
```

```python
# revisar detalles del dataframe
df.info()
```

```python
# Histograma cantidad promedio de habitaciones por ciudad 
fig, ax = plt.subplots()
df2 = df.groupby('TOWN').agg('count')['RM'] # tabla de frecuencia absoluta
ax.bar(df2.index, df2)
plt.show()
```

```python
# Graficos de dispersion "indice de crimen versus Valor medio de la vivienda"
fig, ax = plt.subplots()
# Primero graficaremos el rating versus la cantidad de miembros
ax.scatter(df['CRIM'],df['MEDV'], marker = '*')
ax.set_xlabel('Crimen per capita por ciudad')
ax.set_ylabel('Valor medio de la vivienda USD 1000')
plt.show()

# ¿El indice de crimen afecta al valor medio de la vivienda?
# R: Vemos a mayor indice de criminalidad menor es el valor medio de la vivienda
```

```python
# Graficar los precios medios de las primeras 10 ciudades

# agrupar las primeras 10 ciudades
city = df[:10]['MEDV']

# grafico de barras
city.plot(kind='bar')
```

```python
# Boxplot indice de criminalidad
fig, ax = plt.subplots()
ax.boxplot(df['CRIM'], labels = ['CRIM'])
ax.set_ylabel('Indice de criminalidad')
plt.show()
```
