```python
import pandas as pd
import numpy as np
```

```python
# ejercicio 1

print("ingresar rango de años, rango = mayor - menor")
mayor = int(input("ingrese el año mayor: "))
menor = int(input("ingrese el año menor: "))

if mayor > menor:
  rango = mayor - menor
  
  lista_venta = []
  year = []
  for i in range(rango):
    menor +=i
    year.append(menor)

    venta = float(input("ingresar la ganancia de las ventas del año," + str(year[i])+":"))
    lista_venta.append(venta)

else:
  print("no ingreso los años correctamente")

s = pd.Series(lista_venta)
df = pd.DataFrame()
df["Años"] = year
df["Ventas sin desc"] = lista_venta
df["Ventas 10% desc"] = s*0.9
df

```

```python
# ejercicio 2

# definir la cantidad de alumnos
cantidad = int(input("ingresar la cantidad de alumnos: "))
print("ingresar las notas de los estudiantes")

# definir la funcion informar
def informar(cantidad):
  # crear lista vacia para guardar datos
  list = []
  # ingresar notas
  for i in range(cantidad):
    nota = float(input("ingresar nota: "))
    list.append(nota)
  
  # calcular valores especiales
  min = np.min(list)
  max = np.max(list)
  media = np.median(list)
  desv = np.std(list)
  # crear dos listas
  list_nom = ["minima", "maxima", "media", "desv. estandar"]
  list2 = [min, max, media, desv]
  # construir diccionario a partir de las listas anteriores
  diccionario = dict(zip(list_nom,list2))
  s = pd.Series(diccionario)
  
  return s

informar(cantidad)


```

```python
# ejercicio 3

# ingresar cantidad de alumnos
cantidad = int(input("ingresar la cantidad de alumnos: "))
print("ingresar las notas de los estudiantes")

def ordenar(cantidad):
  # idem ejercicio 3 
  list = []
  for i in range(cantidad):
    nota = float(input("ingresar nota: "))
    list.append(nota)

  s = pd.Series(list)
  
  return s.sort_index(ascending=False)

ordenar(cantidad)
```
