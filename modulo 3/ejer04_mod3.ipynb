```python
# ejercicio e, mostrar por pantalla los nombres de las personas que iban en
# primera clase ordenadas alfabéticamente.
df[df["Pclass"]==1]['Name'].sort_values()
```

```python
# ejercicio f, mostrar por pantalla el porcentaje de personas que sobrevivieron y
# murieron
df['Survived'].value_counts()/df['Survived'].count() * 100
```

```python
# ejercicio g, mostrar por pantalla el porcentaje de personas que sobrevivieron en
# cada clase
df.groupby('Pclass')['Survived'].value_counts(normalize=True)
```

```python
# ejercicio h, eliminar del DataFrame los pasajeros con edad desconocida
df.dropna(subset=['Age'])
```

```python
# ejercicio i, mostrar por pantalla la edad media de las mujeres que viajaban en
# cada clase
df.groupby(['Pclass','Sex'])['Age'].mean().unstack()['female']
```

```python
# ejercicio j, añadir una nueva columna booleana para ver si el pasajero era
# menor de edad o no
df['Young'] = df['Age'] < 18
```

```python
# ejercicio k, mostrar por pantalla el porcentaje de menores y mayores de edad
# que sobrevivieron en cada clase
df.groupby(['Pclass', 'Young'])['Survived'].value_counts(normalize = True) * 100
```
