# Ejercicio 1-2 revisar data

```python
# cargar datos
import pandas as pd
df_abb = pd.read_csv('/content/state-abbrevs.csv',sep=';')
df_area = pd.read_csv('/content/state-area.csv',sep=';')
df_popu = pd.read_csv('/content/state-population.csv',sep=';')

```

# Ejercicio 3 eliminar columnas innecesarias

```python
# limpiar df_area
df_abb_cl = df_abb.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6'], axis=1)
df_abb_cl.head()
```

```python
# limpiar df_area
df_area_cl = df_area.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6'], axis=1)
df_area_cl.head()

```

```python
# limpiar df_popu
df_popu_cl = df_popu.drop(['Unnamed: 4','Unnamed: 5','Unnamed: 6'], axis=1)
df_popu_cl = df_popu.rename(columns = {'state/region':'abbreviation'})
df_popu_cl.head()
```

# Ejercicio 4 agrupar por state/region y ages, para luego agregar usando promedio, mnimo, mximo y contar.

```python
# agrupar por ["state/region","ages"] y calcular promedio
df1 = df_popu_cl.groupby(by = ["abbreviation","ages"]).mean()
df1 = df1.rename(columns = {'population':'mean population'})
df1[["mean population"]]
```

```python
# agrupar por ["state/region","ages"] y calcular el minimo 
df2 = df_popu_cl.groupby(by = ["abbreviation","ages"]).min()
df2 = df2.rename(columns = {'population':'min population'})
df2[["min population"]]
```

```python
# agrupar por ["state/region","ages"] y calcular el maximo 
df3 = df_popu_cl.groupby(by = ["abbreviation","ages"]).max()
df3 = df3.rename(columns = {'population':'max population'})
df3[["max population"]]
```

```python
# agrupar por ["state/region","ages"] y aplicar count()
df4 = df_popu_cl.groupby(by = ["abbreviation","ages"]).count()
df4 = df4.rename(columns = {'population':'count population'})
df4[["count population"]]
```

# Ejercicio 5 formar un nuevo dataset llamado df que tenga la informacin de los 3 sets importados. Use las llaves necesarias.

```python
# merge entre df_abb_cl df_area_cl
df1_full = pd.merge(df_abb_cl,df_area_cl)
df2_full = pd.merge(df1_full,df_popu_cl)
df2_full
```

# Ejercicio 6 calcular la poblacin por rea para cada estado. Para esto agrupe por state/region y calcule el promedio de la poblacin. Guarde dicha informacin en un data frame con las columnas 'state' y 'pop/area' (hint: puede extraer los valores de un objeto groupby, finalizando su sentencia con .values, o extraer sus etiquetas con .index)

```python
df_area_popu = df2_full[["state","area (sq. mi)","population"]]

df_densidad = df2_full["area (sq. mi)"]/df2_full["population"]
df_area_popu["Densidad"] = df_densidad

df_area_popu = df_area_popu.groupby(by = ["state"]).mean()

df_area_popu[["Densidad"]]
```

# Ejercicio 7 crear un dataframe que posea: El estado, el rea, la poblacin, la poblacin por rea y como ndice la abreviacin del nombre del estado.

```python
df = df2_full[["abbreviation","state", "area (sq. mi)", "population"]]

df_densidad = df2_full["area (sq. mi)"]/df2_full["population"]
df["Densidad"] = df_densidad

df
```
